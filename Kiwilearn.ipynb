```python
#Step 1.1: Load the Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Employee_Performance-1.csv')

# Display the first few rows
df.head()
```

```python
#Step 1.2: Summary Statistics
# Summary statistics for numerical columns
df.describe()
```

```python
#Step 1.3: Check for Missing Values
# Check for missing values
df.isnull().sum()
```

```python
#Step 1.4: Data Distribution
# Distribution of Performance Rating
plt.figure(figsize=(8, 4))
sns.histplot(df['Performance Rating'], bins=10, kde=True)
plt.title('Distribution of Performance Rating')
plt.xlabel('Performance Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#Step 1.5: Multivariate Analysis
# Scatter plot of Years of Experience vs. Performance Rating colored by Department
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Years of Experience', y='Performance Rating', hue='Department')
plt.title('Years of Experience vs. Performance Rating by Department')
plt.xlabel('Years of Experience')
plt.ylabel('Performance Rating')
plt.legend(title='Department', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
